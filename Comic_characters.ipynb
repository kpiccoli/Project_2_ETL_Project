```python
import pandas as pd
from sqlalchemy import create_engine
import getpass
import numpy as np
```

### Store CSV into DataFrame

```python
csv_file = "Resources/dc-wikia.csv"
dc_df = pd.read_csv(csv_file)
dc_df.head(10)
```

```python
new_dc_df = dc_df [['page_id', 'name', 'id', 'align', 'alive', 'appearances', 'first appearance', 'year']].copy()
new_dc_df.head()
```

```python
csv_file = "Resources/marvel-wikia.csv"
marvel_df = pd.read_csv(csv_file)
marvel_df.head()
```

```python
new_marvel_df = marvel_df [['page_id', 'name', 'id', 'align', 'alive', 'appearances', 'first appearance', 'year']].copy()
new_marvel_df.head(10)
```

### Transform DataFrame

```python
new_dc_df.dtypes
```

```python
final_dc = new_dc_df.dropna()
final_dc
```

```python
new_marvel_df.dtypes
```

```python
final_marvel = new_marvel_df.dropna()
final_marvel
```

```python
final_marvel['year'] = final_marvel['year'].astype(int)
final_marvel['appearances'] = final_marvel['appearances'].astype(int)
final_marvel
```

```python
#final_marvel['first appearance'] = pd.to_datetime(final_marvel['first appearance'], infer_datetime_format=True)
#final_marvel
```

```python
final_dc = final_dc.rename(columns={'first appearance': 'first_appearance'})
final_dc
```

```python
#sequoyah added
#Splitting name column
dc_heroes = final_dc 

dc_heroes[['super_hero','birth_name']] = dc_heroes.name.str.split("(", n=1, expand=True,) 
dc_heroes 
```

```python
#sequoyah added 
#new table with super hero name, birth name, and year
#dropped the ")" from birth name 
#replaced birth name that appeared as "new eath" with "same"
dc_heroes1 = pd.DataFrame([dc_heroes.super_hero, dc_heroes.birth_name.str.replace(")", "") , dc_heroes.year]).transpose() 
   
dc_heroes1['birth_name'] = dc_heroes1.birth_name.str.replace("new earth", "same") 
dc_heroes1
```

```python
final_marvel = final_marvel.rename(columns={'first appearance': 'first_appearance'})
final_marvel  
```

```python
#sequoyah added
#Splitting name column
marvel_heroes = final_marvel 

marvel_heroes[['super_hero','birth_name']] = marvel_heroes.name.str.split("(", n=1, expand=True,) 
marvel_heroes 
```

```python
#sequoyah added 
#new table with super hero name, birth name, and year
#dropped the ")" from birth name 
#replaced birth name that appeared as "eath 616" with "same"
marvel_heroes1 = pd.DataFrame([marvel_heroes.super_hero, marvel_heroes.birth_name.str.replace(")", "") , marvel_heroes.year]).transpose() 
   
marvel_heroes1['birth_name'] = marvel_heroes1.birth_name.str.replace("(earth-616)", "same") 
marvel_heroes1
```
